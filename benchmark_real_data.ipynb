```python
import pandas as pd
import scanpy as sc

from scipy.stats import rankdata
```

```python
data_with_myc = [
    "FrangiehIzar2021_RNA.h5ad",
    "ReplogleWeissman2022_K562_essential.h5ad",
    "ReplogleWeissman2022_rpe1.h5ad",
]

for file in data_with_myc:
    adata_file = "data/scPerturbData/" + file
    print(adata_file)
    adata = sc.read_h5ad(adata_file)
    print("MYC" in adata.obs['perturbation'].unique().tolist())
```

```python
kale_scores = pd.read_csv("data/_kale_scores_unweighted_TianKampmann2021_CRISPRa_kt_filtered.tsv",
                          sep="\t", index_col=0)
kale_scores = kale_scores.dropna(axis=1, how='all')
kale_scores.fillna(0)

print(kale_scores.shape)
kale_scores.head()
```

```python
viper_scores = pd.read_csv("data/_viper_scores_AdamsonWeissman2016_GSM2406675_10X001_kt_filtered.tsv",
                           sep="\t",
                           index_col=0)
viper_scores = viper_scores.dropna(axis=1, how='all')
print(viper_scores.shape)
viper_scores.head()
```

```python
perturbed_tfs = adata.obs['perturbation'].unique().tolist()
print(len(perturbed_tfs))
perturbed_tfs
```

```python
tfs_in_kale = kale_scores.columns.tolist()

for perturbed_tf in perturbed_tfs:
    if perturbed_tf == 'control' or perturbed_tf == 'nan':
        continue

    if perturbed_tf not in tfs_in_kale:
        continue

    tf_score_df = pd.DataFrame(kale_scores[perturbed_tf], index=kale_scores.index)
    tf_score_df = tf_score_df.sort_values(by=perturbed_tf, ascending=True)

    # Get cells perturbed for this TF
    marked_cells = adata.obs[adata.obs['perturbation'] == perturbed_tf].index
    print(f"{perturbed_tf} {"-" * 30}")
    print(f"\tcells perturbed: {len(marked_cells)}")

    tf_score_df["marked"] = False
    tf_score_df.loc[marked_cells, "marked"] = True

    # Rank the scores between 0 and 1
    tf_score_df['rank'] = (rankdata(tf_score_df[perturbed_tf], method='average') - 0.5) / len(tf_score_df)

    # Calculated mean rank of marked cells
    mean_rank_marked = tf_score_df[tf_score_df['marked'] == True]['rank'].mean()
    print(f"\tMean rank: {mean_rank_marked}")
    # Also print how many targets these TF have in the network

    # Sort them ascending
    # Find n=total cells
    # Find k=number of marked cells
    # find sd analytically, and mean = 0.5
    # Calculate cdf which is also the p-value, this is because these are inhibited, otherwise we would do (1-cdf)
    # Confirm this is significant for MYC (Before find dataset where MYC is perturbed)
```
